```python
import pandas as pd
import re

ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
joined = ratings.merge(movies, how='left', on='movieId')
joined.head()
```

```python
joined['year'] =joined['title'].str.extract('.*\((.*)\).*',expand = False)
joined.loc[pd.isnull(joined.year), 'year'] = '1900'
joined.head(5)
```

```python
#Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ 
#и отсортируйте результат по убыванию рейтинга

avg_rating = joined.groupby('year').agg({'rating': 'mean'}).sort_values('rating', ascending=False).head()
avg_rating
```

```python
import pandas as pd
import re

ratings_ = pd.read_csv('ratings.csv')
movies_ = pd.read_csv('movies.csv')
joined_ = ratings_.merge(movies_, how='left', on='movieId')
joined_.head()
```

```python
years_ = [str(i) for i in range(1950, 2011)]

def production_year(movie):
    for y in years_:
        if y in movie:
            return y
        
    return '1900'  
        
joined_['years'] = joined_['title'].apply(production_year)
joined_.head()

avg_rating_ = joined_.groupby('years').agg({'rating': 'mean'}).sort_values('rating', ascending=False).head()
avg_rating_
```

```python
data_r = pd.read_csv('ratings.csv')
data_r.head()
```

```python
result = data_r.groupby('userId').agg({
    'timestamp': ['max', 'min', 'count']
}).reset_index()
result.head()
```

```python
def fans(rating):
    if rating > 100:
        return True
    return False

result['more_than_100'] = result['timestamp']['count'].apply(fans)

result_ = result[result['more_than_100']]
                 
life = result_['timestamp']['max'] - result_['timestamp']['min']
life.max()
```
