```python
import pandas as pd
```

```python
data_m = pd.read_csv('ratings.csv')
data_m.head()
```

```python
def rating_estimate(param):
    if param <= 2:
        return 'Низкий рейтинг'
    if param <= 4:
        return 'Средний рейтинг'
    if param <= 5:
        return 'Высокий рейтинг'
```

```python
data_m['class'] = data_m['rating'].apply(rating_estimate)
data_m.head()
```

```python
data_geo = pd.read_csv('keywords.csv')
data_geo.head(5)
```

```python
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
'Северо-Запад': ['петербург', 'псков', 'мурманск'],
'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']}
```

```python
def geo(param):
    for v in geo_data.values():
        if v[0] in data_geo:
            return 'Центр'
        if v[1] in data_geo:
            return 'Северо-Запад'
        if v[2] in data_geo:
            return 'Дальний Восток'
    return 'undefined'
```

```python
data_geo['region'] = data_geo['keyword'].apply(geo)
data_geo.head()
```
